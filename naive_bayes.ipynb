```python
# Naive Bayes Classifier
```

```python
# Import Libraries
from sklearn.naive_bayes import BernoulliNB, GaussianNB, MultinomialNB
from sklearn.datasets import load_digits, load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Bernoulli Model
data = load_digits()
X = data.data
y = data.target

# Convert features to binary (boolean) features
X_binary = (X > 0).astype(int)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_binary, y, test_size=0.3, random_state=42)

# Initialize Bernoulli Naive Bayes classifier
bnb = BernoulliNB()

# Train the classifier
bnb.fit(X_train, y_train)

# Predict on the test set
y_pred = bnb.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f'Bernoulli Naive Bayes Accuracy: {accuracy:.2f}')
```

```python
# Gaussian Model
data = load_iris()
X = data.data
y = data.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize Gaussian Naive Bayes Classifier
gnb = GaussianNB()

# Train the classifier
gnb.fit(X_train, y_train)

# Predict on the test set
y_pred = gnb.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f'Gaussian Naive Bayes Accuracy: {accuracy:.2f}')
```

```python
# Multinomial Model
data = load_digits()
X = data.data
y = data.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize Multinomial Naive Bayes Classifier
mnb = MultinomialNB()

# Train the classifier
mnb.fit(X_train, y_train)

# Predict on the test set
y_pred = mnb.predict(X_test)

# Evaluate the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f'Multinomial Naives Bayes Accuracy: {accuracy:.2f}')
```

```python
# Plotting the data distributions of each data used in the various Naive Bayes Classifier
```

```python
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Bernoulli Model
# Load the Digits dataset
digits_data = load_digits()
X_digits = digits_data.data
y_digits = digits_data.target

# Convert pixel intensities to binary (presence or absence)
X_binary = (X_digits > 0).astype(int)

# Create a DataFrame for binary features
binary_digits_df = pd.DataFrame(data=X_binary)

# Plot histograms for binary features
plt.figure(figsize=(12, 6))
for i in range(X_binary.shape[1]):
    plt.hist(binary_digits_df[i], bins=2, alpha=0.5, label=f'Feature {i}')

plt.title('Histogram of Binary Pixel Features in Digits Dataset')
plt.xlabel('Feature Value')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

```python
# Gaussian Model
# Load the Iris dataset
iris_data = load_iris()
X_iris = iris_data.data
y_iris = iris_data.target
iris_df = pd.DataFrame(data=X_iris, columns=iris_data.feature_names)
iris_df['species'] = iris_data.target_names[y_iris]

# Plot distributions for each feature by species
sns.pairplot(iris_df, hue='species', diag_kind='kde')
plt.suptitle('Kernel Density Estimation of Iris Dataset Features by Species', y=1.02)
plt.show()
```

```python
# Multinomial Model

# Load the Digits dataset
digits_data = load_digits()
X_digits = digits_data.data
y_digits = digits_data.target

# Plot histograms for pixel intensities
plt.figure(figsize=(12, 6))
plt.hist(X_digits.ravel(), bins=30, color='blue', edgecolor='black')
plt.title('Histogram of Pixel Intensities in Digits Dataset')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
